# DATA CLEANING: cleaning the data set 

```python
cars=cars.rename(columns={'Unnamed :1': ' model'})
print(cars)
```

```python
cars.qsec=cars.qsec.fillna(cars.qsec.mean())
cars
```

```python
# drop the unwanted column 
cars=cars.drop(columns=['disp'])
cars
```

```python
#changing the datatype 
cars.disp=cars.disp.astype(float)
cars
```
