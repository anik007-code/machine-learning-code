```python
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv("iris.csv")
#print(df)
df=df.values
```

```python
X=df[:,0:4]
Y=df[:,4]
#print(X,Y)
```

```python
l=[]
for i in range(Y.shape[0]):
    if(Y[i]==" Iris-setosa"):
        l.append(0)
    elif(Y[i]=="Iris-virginica"):
        l.append(1)
    else:
        l.append(2)
#print(l)        
```

```python
plt.subplot(7,1,1)
plt.scatter(X[:,0],X[:,1],c=l)
plt.subplot(7,1,2)
plt.scatter(X[:,0],X[:,2],c=l)
plt.subplot(7,1,3)
plt.scatter(X[:,0],X[:,3],c=l)
plt.subplot(7,1,4)
plt.scatter(X[:,1],X[:,2],c=l)
plt.subplot(7,1,5)
plt.scatter(X[:,1],X[:,3],c=l)
plt.subplot(7,1,1)
plt.scatter(X[:,2],X[:,3],c=l)
plt.show()
```

```python
from sklearn.model_selection import train_test_split 
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.svm import SVC
C=SVC()
C.fit(xtrain, ytrain)
```

```python
y_pred=C.predict(xtest)
print(y_pred.shape)
```

```python
import numpy as np
print(np.sum(np.where(y_pred!=ytest,1,0)))
```
